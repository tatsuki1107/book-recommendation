# Matrix Foctorizationの挙動を確認する（UI上で確認すると同じような本しか推薦されていない）

## 本の数が1121冊もあるが、推薦リストに含まれていた本の数は107冊だけの理由を考察

・　そもそもユーザの好みが似ている  
・　学習時にはuser数、本の数を絞っていたので、そのことに関するバイアスが働いている

## そのことに関する欠点

・　多様な推薦ができていない（厳密には、dateのカラムが存在しないので、「ユーザが好むであろう本を推薦したとき、ユーザが好む割合を高くする（presicion）」モデルのもとで多様性がない）  
・　このプラットフォームに参加している出版社、著者に対して不公平。（同じ商品ばかりが推薦されるから）

```python
import joblib

user_df, book_df, rating_df = joblib.load('../data/mf-api.pkl')
```

```python
print(f"ユーザー数: {len(user_df)}")
print(f'本の数: {len(book_df)}')
print(f'評価数: {len(rating_df)}')
```

```python
import numpy as np
from itertools import chain

recommended_books = user_df["mf_recommend_item"].to_list()
recommended_books = list(chain.from_iterable(recommended_books))
```

```python
book_count = {}
for book in recommended_books:
    book_count[book] = 1 + book_count.get(book, 0)
```

```python
print(f"各ユーザにレコメンドされた本のユニーク数: {len(book_count)}")
```

```python
book_count = dict(sorted(book_count.items(), key=lambda x:x[0]))
```

```python
rec_df = rating_df[rating_df["book_id"].isin(set(recommended_books))].groupby("book_id").agg({"book_rating": ["mean", "count"]}).sort_values(by="book_id")
rec_df["recommend_count"] = list(book_count.values())
```

```python
rec_df.sort_values(by="recommend_count", ascending=False, inplace=True)
```

```python
import seaborn as sns; sns.set()
import matplotlib.pyplot as plt
%matplotlib inline

fig, (ax1, ax2) = plt.subplots(1,2, figsize=(20,5))

sns.scatterplot(rec_df, x=("book_rating", "count"), y="recommend_count", ax=ax1)
sns.scatterplot(rec_df, x=("book_rating", "mean"), y="recommend_count", ax=ax2)
```

# rating付与された数が少なく、かつ高いratingのアイテムに過学習している！！！
  
・　学習するアイテム数に対して相対的に評価された数が少ないと、過学習になるよう。なので、正則化項を強めるのが解決策？

```python
over_learning_count = rec_df[rec_df["recommend_count"]> 250]["recommend_count"].values.sum() 
k = 5
print(f'過学習しているアイテムが全推薦リストに含まれている割合: {over_learning_count / (k*len(user_df)):.2f}')
```

```python
# 過学習していないアイテムだけ見てみる
fig, (ax1, ax2) = plt.subplots(1,2, figsize=(20,5))

sns.scatterplot(rec_df[rec_df["recommend_count"]<250], x=("book_rating", "count"), y="recommend_count", ax=ax1)
sns.scatterplot(rec_df[rec_df["recommend_count"]<250], x=("book_rating", "mean"), y="recommend_count", ax=ax2)
```

# 外れ値を除いても傾向は変わらない（それはそう）
### つまりこのMFは、過学習したアイテムを9割推薦モデル。つまり、評価数が少なく、高評価なアイテムを9割推薦するモデルであるので、「機械学習」を使う必要のないものをわざわざ実装したということになる
