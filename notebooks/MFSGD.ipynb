# 確率的勾配下降法で過学習を確認

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('ggplot')
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
import joblib
import sys; sys.path.append("..")
from src.MFSGD import MatrixFactorization
#from multiprocessing import Pool, cpu_count
```

```python
train_df, test_df = joblib.load('../data/train_test_split.pkl')
```

```python
train_df["book_rating"] = train_df["book_rating"].astype('int64')
test_df["book_rating"] = test_df["book_rating"].astype('int64')
```

```python
user_df = pd.DataFrame(train_df["user_index"].sort_values().unique())
user_id2_index = dict(zip(user_df[0].to_list(), user_df.index.to_list()))
```

```python
book_df = pd.DataFrame(train_df["book_index"].sort_values().unique())
book_id2_index = dict(zip(book_df[0].to_list(), book_df.index.to_list()))
```

```python
ui, bi = [], []
for user_index, book_index in zip(train_df["user_index"].to_list(), train_df["book_index"].to_list()):
    ui.append(user_id2_index[user_index])
    bi.append(book_id2_index[book_index])
    
train_df["user_index"] = ui
train_df["book_index"] = bi
    
```

```python
ui, bi = [], []
for user_index, book_index in zip(test_df["user_index"].to_list(), test_df["book_index"].to_list()):
    ui.append(user_id2_index[user_index])
    bi.append(book_id2_index[book_index])
    
test_df["user_index"] = ui
test_df["book_index"] = bi
```

```python
train = train_df.to_numpy()
test = test_df.to_numpy()
```

```python
train, val = train_test_split(train, test_size=0.30, random_state=12345)
```

```python
#p = Pool(processes=cpu_count()-1)

mf = MatrixFactorization(
    k=100,
    learning_rate=2.5,
    reg_param=0.5,
    random_state=12345,
    alpha=1.0,
    #beta1=1.0,
    #beta2=1.0,
    eps=0.7
)
train_loss, val_loss, test_loss = mf.fit(
    train=train, val=val, test=test, n_epochs=50
)
```

## 学習曲線

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20,5))
ax1.plot(np.arange(len(train_loss)), train_loss, label="train RMSE", linewidth=3, linestyle="dashed")
ax1.plot(np.arange(len(val_loss)), val_loss, label="val RMSE", linewidth=3)
#plt.plot(np.arange(len(test_loss)), test_loss, label="test RMSE", linewidth=3)
   
ax1.set_title("Train/Validation/test Curves", fontdict=dict(size=20))
ax1.set_xlabel("Number of Epochs", fontdict=dict(size=20))
ax1.set_ylabel("Root Mean Squared Error", fontdict=dict(size=20))
ax1.set_ylim([0.0, 3.0])
ax1.legend(loc="best", fontsize=10)

matrix = mf.P@mf.Q.T
my_cmap = plt.cm.get_cmap('Reds')
heatmap = plt.pcolormesh(matrix.T, cmap=my_cmap)
colorbar = plt.colorbar(heatmap, ax=ax2)
ax2.grid()
plt.tight_layout()
plt.show()
```
